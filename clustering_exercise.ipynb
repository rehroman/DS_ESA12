```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import pandas

X = pandas.read_csv('./edlich-kmeans-A0.csv')

```

```python
X.head()
```

```python
# EINSTELLEN DER CLUSTER-ANZAHL (k) --> 4 optimal
kmeans = KMeans(n_clusters=4)
#kmeans = KMeans(n_clusters=10)

kmeans = kmeans.fit(X)

labels = kmeans.predict(X)

C = kmeans.cluster_centers_
```

```python
print(C)
```

```python
plt.scatter(X['V1'], X['V2'], X['V3'])
```

```python
# Zuordnung der Werte zu Clustern vornehmen
y_km = kmeans.fit_predict(X)
```

```python
X['cluster'] = y_km
# Zuordnung der Cluster anzeigen
X.head()
```
